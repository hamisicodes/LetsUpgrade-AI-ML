```python
import pandas as pd
import  numpy as np
```

```python
titanic_train = pd.read_csv('train.csv')
titanic_train.head()
```

```python
titanic_train.isna().sum()
```

```python
titanic_train.dtypes
```

```python
titanic_train['Age'] 
```

```python
titanic_train['Fare'] 
```

```python
titanic_train['Age'] = titanic_train['Age'].astype(int)
titanic_train['Fare'] = titanic_train['Fare'].astype(int)
```

```python
titanic_train['Age'] 
```

```python
titanic_train['Fare']
```

```python
titanic_train.describe().T
```

```python
new_age_var = np.where(titanic_train['Age'].isna(), 32 ,titanic_train['Age'] )
titanic_train['Age'] = new_age_var
```

```python
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()

titanic_train['Sex'] = label_encoder.fit_transform(titanic_train['Sex'])
```

```python
from sklearn import tree

tree_model = tree.DecisionTreeClassifier()

predictors = pd.DataFrame([titanic_train['Sex'],titanic_train['Age'],titanic_train['Fare']]).T

tree_model.fit(X = predictors , y = titanic_train['Survived'])
```

```python
with open('Dtree.dot', 'w') as f:
    f = tree.export_graphviz(tree_model , feature_names=['Sex','Age','Fare'], out_file=f)
```

```python
# model accuracy

tree_model.score(X= predictors , y= titanic_train['Survived'])
```

```python
# The model accuracy for classification is 94%
```

### Predicting using the built model

```python
# load test dataset
titanic_test = pd.read_csv('test.csv')
titanic_test.head()
```

```python
titanic_test.isna().sum()
```

```python
# encoding the sex column
titanic_test['Sex'] = label_encoder.fit_transform(titanic_test['Sex'])
```

```python
titanic_test['Age'] = titanic_test['Age'].astype(int)
titanic_test['Fare'] = titanic_test['Fare'].astype(int)
```

```python
titanic_test
```

```python
test_features = pd.DataFrame([ titanic_test['Sex'],titanic_test['Age'],titanic_test['Fare'] ]).T

test_preds = tree_model.predict(X=test_features)

predicted_output = pd.DataFrame({'PassengerID':titanic_test['PassengerId'],'Survived':test_preds})

predicted_output.to_csv('titanic_test_predictions.csv' , index =False);
```
