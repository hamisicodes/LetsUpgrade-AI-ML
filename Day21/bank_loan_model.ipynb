```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_excel('Bank_Personal_Loan_Modelling.xlsx' , sheet_name=1)
dataset.head()
```

```python
dataset.isna().sum()
```

### Drop zipcode and ID variables

```python
dataset.drop(['ZIP Code','ID'] , axis = 1 , inplace=True)
```

```python
dataset
```

### split the variables into x and y

```python
# Dependent/Target variable
y = dataset['Personal Loan']
y.head()
```

```python
# Independent variables 
x = dataset[['Age','Experience','Income' ,'Family' , 'CCAvg' , 'Education' , 'Mortgage' , 'Securities Account' ,'CD Account' , 'Online' , 'CreditCard']]
x.head()
```

```python
import statsmodels.api as sm

x1 = sm.add_constant(x)
Logistic = sm.Logit(y,x1)
result = Logistic.fit()
result.summary()
```

### inferences:
* Significant Independent variables ie p-value is less than 0.05 include:
    1. Income
    1. Family
